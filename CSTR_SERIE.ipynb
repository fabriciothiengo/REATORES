## **REATORES CSTR EM SÉRIE**
---

Sejam 2 reatores CSTR em série de mesmo volume, alimentados com vazão volumétrica (w) de 1 $m^3/h$ e k=3,8 $h^{-1}$. Com $C_{A_0}$ = 2,3 mol/l e  $X_2$ = 0,8, qual deve ser $X_1$ para minimizar o volume total?

O volume para o primeiro tanque será:

$$
V_1=\frac{F_{A_0}X_1}{kC_{A_0}(1-X_1)}
$$

Para o segundo tanque o volume será:
$$
V_2 = \frac{F_{A_0}(X_2-X_1)}{kC_{A_0}(1-x_2)}
$$

O volume total será a soma dos volumes

$$
V=\frac{F_{A_0}}{kC_{A_0}}\left(\frac{X_1}{1-X_1}+\frac{X_2-X_1}{1-X_2}\right)
$$

Existem pelo menos duas formas de determinar qual valor de $X_1$ minimiza o volume total.

#### **SOLUÇÃO 1: DERIVAR A EXPRESSÃO EM FUNÇÃO DE $X_1$ E ENCONTRAR O ZERO DA FUNÇÃO**

A derivada da função de V em relação a $X_1$ é:

$$
\frac{dV}{dX_1}=\frac{F_{A_0}}{kC_{A_0}}\left[\frac{1}{(1-X_1)^2}-\frac{1}{1-X_2}\right]
$$

```python
import numpy as np
from scipy.optimize import fsolve

# parâmetros (ajuste conforme seu problema)
w=1.0
k = 3.8
C = 1.0
F=C*w
X2 = 0.8  # fixo

# derivada dV/dX1
def dVdX1(X1):
    return (F/(k*C)) * (1/(1 - X1)**2 - 1/(1 - X2))

X1_sol = fsolve(dVdX1, 0.5)

print("X1 ótimo =", f"{X1_sol[0]:.2f}")
```

#### **SOLUÇÃO 2: DETERMINAR O VOLUME MÍNIMO UTILIZANDO O MÉTODO DO ENXAME DE PARTÍCULAS**

```python
! pip install pyswarm --q
import numpy as np
from pyswarm import pso

# parâmetros
w=1.0
k = 3.8
CA0 = 1.0
FA0=CA0*w
X2 = 0.8  # fixo

# função objetivo (PSO sempre minimiza)
def fobj(X):
    X1 = X[0]
    return (FA0/(k*CA0)) * (X1/(1 - X1) + (X2 - X1)/(1 - X2))

# rodar PSO
Xopt, Vopt = pso(fobj, [1e-6], [X2-1e-6])

print(f'O volume ótimo de {Vopt:.2f} litros ocorre em X1 = {Xopt[0]:.2f}')
```
