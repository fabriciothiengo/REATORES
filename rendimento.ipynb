**AVALIAÇÃO DO RENDIMENTO EM REATORES CONTÍNUOS**

---
Sejam as reações em parelelo

<br>
$A → B $ / $r_B$ = 2

$A → C $ / $r_C$ = 2$C_A^2$

$A → D $ / $r_C$ = 0,72$C_A^{2,5}$

<br>
O rendimento do produto C é dado por:

$Y_C=\frac{r_C}{r_B+r_C+r_D}$ = $\frac{2C_A^2}{2+2C_A^2+0,72C_A^{2,5}}$

---
Outro raciocínio que podemos fazer é sobre o quanto de produtos desejado podemos obter em função da quantidade de reagente alimentado, multiplicando o rendimento (produto desejado/reagente consumido) pela conversão (reagente consumido / reagente alimentado). O desejável então seria maximizar esse parâmetro.


Qual seria então a condição de conversão que maximiza o rendimento global para o reator CSTR e para o PFR, considerando uma concentração de alimentação $C_{A_0}$ = 1,5 M.

```python
import numpy as np
from scipy.integrate import cumulative_trapezoid
import matplotlib.pyplot as plt

CA0 = 1.5

X = np.linspace(0, 0.95, 100)

YC = 2*CA0**2*(1-X)**2/(2+2*CA0**2*(1-X)**2+0.72*CA0**2.5*(1-X)**2.5)

I = cumulative_trapezoid(YC, X, initial=0)
YP=(1/X)*I

razao=YP/YC

PC=YC*X
PP=YP*X

plt.figure(1)
plt.plot(X, YC, label='CSTR')
plt.plot(X, YP, label='PFR')
plt.xlabel('conversão')
plt.ylabel('rendimento global')
plt.legend()

plt.figure(2)
plt.plot(X[1:-2], razao[1:-2])
plt.xlabel('conversão')
plt.ylabel('razão de rendimento')

plt.figure(3)
plt.plot(X, PC, label='CSTR')
plt.plot(X, PP, label='PFR')
plt.xlabel('conversão')
plt.ylabel('produto desejado por reagente alimentado')
plt.legend()
```
