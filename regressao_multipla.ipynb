# Regressão Linear Múltipla

A regressão linear múltipla é utilizada para modelar a relação entre uma variável dependente y e múltiplas variáveis independentes xi.

---

## Modelo Matemático

Para a reação $$ A + 2B → R + S$$
A expressão de taxa será $$r_A=-kC_A^aC_B^b$$

A expressão da taxa de reação após transformação linear fica:

$$
ln\left(-\frac{dC_A}{dt}\right) = lnk + a \ lnC_A + b \ lnC_B
$$

Essa expressão da taxa equivale ao seguinte modelo linear

O modelo é dado por:

$$
y = a_0 + a_1 x_1 + a_2 x_2
$$

Sendo $$x_0 = 1$$ $$x_1 = ln C_A$$ $$x_2 = ln C_B$$

A solução genérica é na forma $$y=Xa$$

Na matriz X, cada coluna é preenchida com os valores $x_i$, sendo que na primeira coluna todos os valores são 1.

A solução por mínimos quadrados é:

$$a = \left(X^TX\right)^{-1}X^Ty$$


##**Aplicação**
---

Temos os seguintes dados de tempo e concentração:

$$t=[0,2,4,6,8,10,15,20,25,30,50,70,100,200,600]$$

$$X=[0,0.17,0.28,0.38,0.43,0.5,0.6,0.68,0.71,0.75,0.85,0.87,0.91,0.95,0.98]$$

Como a relação estequiométrica A:B é 1:2 e $C_{A_0}$:$C_{B_0}$ também é 1:2, tanto faz quem é o limitante. Assim, podemos escrever:

$$C_A = C_{A_0}(1-X)$$
$$C_B = C_{A_0}(2-2X)$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Dados de exemplo
t=np.array([0,2,4,6,8,10,15,20,25,30,50,70,100,200,600])
X=np.array([0,0.17,0.28,0.38,0.43,0.5,0.6,0.68,0.71,0.75,0.85,0.87,0.91,0.95,0.98])
CA=0.5*(1-X)
CB=2*CA
x1 = np.log(CA)
x2 = np.log(CB)

dCdt=np.zeros(len(t))

dCdt[0]=(CA[1]-CA[0])/(t[1]-t[0])

for i in range(1,len(t)-1):
  dCdt[i]=(CA[i+1]-CA[i-1])/(t[i+1]-t[i-1])

dCdt[-1]=(CA[-1]-CA[-2])/(t[-1]-t[-2])

y  = np.log(-dCdt)

# Montar matriz X (com coluna de 1 para o intercepto)
X = np.column_stack((np.ones(len(x1)), x1, x2))

# Cálculo dos coeficientes (mínimos quadrados)
#a = np.linalg.inv(X.T @ X) @ X.T @ y
a = np.linalg.pinv(X) @ y

print("Coeficientes:")
print(f"k = {np.exp(a[0]):.4f}")
print(f"a = {a[1]:.4f}")
print(f"b = {a[2]:.4f}")
```
