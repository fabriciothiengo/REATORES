**INFLUÊNCIA DA TEMPERATURA E DA ENERGIA DE ATIVAÇÃO SOBRE A VELOCIDADE ESEPECÍFICA DAS REAÇÕES**

```python
import numpy as np
import matplotlib.pyplot as plt

T=np.linspace(300,3000,100)
R=8.314
k0=10**10
Ea=np.array([20,50,100])*10**3
k=np.empty((len(T),len(Ea)))

for j in range(len(Ea)):
  for i in range(len(T)):
    k[i,j]=k0*np.exp(-Ea[j]/(R*T[i]))


plt.plot(T-273,k,label=(['20','50','100']))
plt.legend()
plt.title('INFLUÊNCIA DA TEMPERATURA E ENERGIA DE ATIVAÇÃO SOBRE A VELOCIDADE ESPECÍFICA')
plt.xlabel('Temperatura (oC)')
plt.ylabel('Velocidade específica')
```

Como pode ser visto no gráfico, a energia de ativação tem uma forte
influência sobre a velocidade específica. Percebe-se que reações com baixa energia de ativação são favorecidas já em baixas temperatura, enquanto que reações com alta energia ocorrem com velocidade significativa apenas em temperatura elevada, o que pode não ser viável. Esse é o caso clássico de se justificar a utilização de catalisadores, pois as reações com alta energia ocorrem com velocidade relativamente baixa, resultando em baixa produtividade. Ao se utilizar o catalisador, a energia é reduzida e a velocidade aumenta significativamente e, por consequência, a produtividade. Embora haja um custo associado ao uso do catalisador, o ganho na produtividade compensa, e muito, o investimento com o catalisador.
