## **REAÇÃO EM BATELADA A VOLUME VARIÁVEL**
---

Seja a reação elementar em fase gasosa ocorrendo em um reator batelada (A + B $\rightarrow$ C). As concentrações iniciais de A e B são, respectivamente, 1M e 2M. A velocidade específica da reação é 2,5 l/mol h.
Qual a conversão depois de 15 minutos de reação?

---

O balanço de massa para uma reação em batelada a volume variável é o seguinte:

$$
\frac{d(C_AV)}{dt}=C_A\frac{dV}{dt} + V\frac{dC_A}{dt} = - kC_AC_BV
$$

$$
C_A=\frac{C_{A_0}(1-X)}{1+ \epsilon X}
$$

$$
C_B=\frac{C_{A_0}(2-X)}{1+ \epsilon X}
$$

$$
dC_A = \frac{-C_{A_0}(1+\epsilon X)}{(1+\epsilon X)^2}dX
$$

$$
V=V_0(1+\epsilon X)
$$

$$
dV=V_0 \epsilon dX
$$

$$
y_{A_0} = 1/3 \\ \delta=-1 \\ \epsilon = y_{A_0} \cdot \delta = -1/3
$$

Inserindo todos os termos no balanço de massa e agrupando os termos semelhantes, chega-se a:

$$
(1+\epsilon X)\frac{dX}{dt}=kC_{A_0}(1-X)(2-X)
$$

Nesse caso, existem duas soluções possíveis

## **SOLUÇÃO: RESOLVENDO A EQUAÇÃO NA FORMA DIFERENCIAL**

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Parâmetros
k = 2.5          # constante cinética
CA0 = 1.0        # concentração inicial
eps = -1/3        # epsilon

# Função da EDO
def dXdt(t, X):
    return (k * CA0 * (1 - X) * (2 - X)) / (1 + eps * X)

# Condição inicial
X0 = [0]

# Intervalo de tempo
tspan = [0, 0.25]
t_eval = np.linspace(tspan[0], tspan[1], 100)

# Resolução
sol = solve_ivp(dXdt, tspan, X0, t_eval=t_eval, method='RK45')

# Resultado
t = sol.t
X = sol.y[0]

# Plot
plt.plot(t, X)
plt.xlabel('tempo (h)')
plt.ylabel('X')
plt.grid()
plt.show()
```
